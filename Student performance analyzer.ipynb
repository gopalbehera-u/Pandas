# Data exploration

```python
import pandas as pd
import numpy as np

data = {
    'Name':    ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Henry'],
    'Gender':  ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'M'],
    'Math':    [85, 62, 90, 45, 78, 55, 92, 38],
    'English': [78, 70, 88, 50, 82, 60, 95, 42],
    'Science': [80, 65, 92, 48, 75, 58, 89, 40],
    'Attend':  [90, 75, 95, 60, 88, 70, 98, 55]
}

df = pd.DataFrame(data)
```

```python
print(df.shape)
print(df.head(5))
print(df.info())
print(df.describe())
```

```python
df.head()
```

#  Student wise analysis

```python
df
```

```python
df['Total']=df[['Math','English','Science']].sum(axis=1)
df['Average']=df[['Math','English','Science']].mean(axis=1)
df['Grade']=np.where(df['Average']>=80,'A',np.where(df['Average']>=60,'B',np.where(df['Average']>=40,'C','F')))

df[['Name','Total','Average','Grade']]
```

# Subject wise analysis


```python
df[['Math','English','Science']].mean()
avg=df[['Math','English','Science']].mean()
print(avg.idxmax())
idx=df['Math'].idxmax()
print(df['Name'][idx],df['Math'][idx])
idx=df['English'].idxmax()
print(df['Name'][idx],df['English'][idx])
idx=df['Science'].idxmax()
print(df['Name'][idx],df['Science'][idx])
print((df['Math']>80).sum())
print((df['English']>80).sum())
print((df['Science']>80).sum())
```

# Gender wise analysis

```python
print(df.groupby('Gender')[['Math','Science','English']].mean())
print(df.groupby('Gender')['Average'].mean().idxmax())
print(df['Gender'].value_counts())
```

# Attendance analysis

```python
# task 1 — average attendance
print(df['Attend'].mean())

# task 2 — below 75 attendance
print(df[df['Attend'] < 75]['Name'])

# task 3 — correlation between attendance and average
print(df['Attend'].corr(df['Average']))
# close to 1 = strong positive correlation
# close to 0 = no correlation

# task 4 — classify attendance
df['Attend_Status'] = np.where(df['Attend'] >= 90, 'Excellent',
                      np.where(df['Attend'] >= 75, 'Good', 'Poor'))
print(df[['Name', 'Attend', 'Attend_Status']])
```
