# Dataset

```python
import pandas as pd
import numpy as np

data={

'OrderID': [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010],
'Product': ['Laptop','Phone','Tablet','Laptop','Watch',
'Phone','Tablet','Laptop','Watch','Phone'],
'Category': ['Electronics','Electronics','Electronics','Electronics','Accessories',
'Electronics','Electronics','Electronics','Accessories','Electronics'],
'Quantity': [2, 5, 3, 1, 8, 4, 2, 3, 6, 7],
'Price': [75000,25000,35000,75000,15000,25000,35000,75000,15000,25000],
'Region': ['North','South','East','West','North','East','South','North','West','South']
,
'SalesPerson':['Alice','Bob','Charlie','Alice','David','Bob','Charlie','David','Alice','Bob'],
'Month': ['Jan','Jan','Jan','Feb','Feb','Mar','Mar','Mar','Feb','Jan']
}
```

```python
df=pd.DataFrame(data)
```

# Data exploration

```python
print(df.shape)
print(df.head())
print(df.info())
print(df.describe())
print(df['Product'].unique())
print(df['Region'].unique())
print(df['Region'].value_counts())
```

# Revenue analysis


```python
df['Revenue']=df['Quantity']*df['Price']
print(df.head(2))
revenue_per_region=df.groupby('Region')['Revenue'].sum()
print(revenue_per_region)
revenue_per_order=df.groupby('OrderID')['Revenue'].sum().mean()
print(revenue_per_order)
top_orders = (
    df.groupby('OrderID')['Revenue'].sum().sort_values(ascending=False).head(3)
)

print(top_orders)
```

#  Product analysis

```python
revenue_per_product=df.groupby('Product')['Revenue'].sum()
print(revenue_per_product)
best_selling_product=(df.groupby('Product')['Quantity'].sum().idxmax())
print(best_selling_product)
average_price_category=df.groupby('Category')['Price'].mean()
print(average_price_category)
high_revenue=df.groupby('Product')['Revenue'].sum().idxmax()
print(high_revenue)
```

#  Region & salesperson analysis

```python
total_revenue_region=df.groupby('Region')['Revenue'].sum()
print(total_revenue_region)
total_revenue_region_best=df.groupby('Region')['Revenue'].sum().idxmax()
print(total_revenue_region_best)
total_revenue_salesperson=df.groupby('SalesPerson')['Revenue'].sum()
print(total_revenue_salesperson)
total_revenue_salesperson_top=df.groupby('SalesPerson')['Revenue'].sum().idxmax()
print(total_revenue_salesperson_top)
```

#  Monthly trend

```python
total_revenue_month=df.groupby('Month')['Revenue'].sum()
print(total_revenue_month)
total_revenue_month_high=df.groupby('Month')['Revenue'].sum().idxmax()
print(total_revenue_month_high)
total_number_oforder_permonth=df.groupby("Month")['OrderID'].count()
print(total_number_oforder_permonth)
```

# Filtering & classification

```python
revenue_above_100k=df[df['Revenue']>100000]
print(revenue_above_100k)
electroninc_from_north=df[(df['Region']=='North') & (df['Category']=='Electronics')]
print(electroninc_from_north)
df['Revenue_Class']=np.where(df['Revenue']>200000,'High',np.where(df['Revenue']>100000,'Medium','Low'))
print(df['Revenue_Class'])
```
